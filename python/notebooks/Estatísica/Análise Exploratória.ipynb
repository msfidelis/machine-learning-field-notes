## DistribuiÃ§Ã£o dos Dados

### DistribuiÃ§Ã£o Normal

#### Problema HipotÃ©tico

> Qual a probabilidade de uma emissÃ£o de plastico ser menor que 4%?



> Qual a probabilidade de uma emissÃ£o de palstico ser maior que 60%?


```python
stats.describe(df.composition_plastic_percent)
```

#### Encontrando o Desvio PadrÃ£o da EmissÃ£o de Plastico

```python
plastic_std_variation = np.std(df.composition_plastic_percent)
plastic_std_variation
```

#### Encontrando a mÃ©dia de emissÃ£o de plastico

```python
plastic_mean = np.mean(df.composition_plastic_percent)
plastic_mean
```

#### Fazendo o teste de probabilidade e encontrar um valor menor que 4

```python
test_plastic = 4
```

```python
stats.norm.cdf(test_plastic, plastic_mean, plastic_std_variation)
```

#### Fazendo o teste de probabilidade de encontrar um valor maior que 60%

```python
test_plastic = 60
```

```python
stats.norm.sf(test_plastic, plastic_mean, plastic_std_variation)
```
