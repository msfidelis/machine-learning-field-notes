```python
import pandas as pd 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

## Brazil Gas Prices - Exploratory Analysis

```python
df = pd.read_csv("datasets/combustiveis-brasil.csv")
```

```python
fig, ax = plt.subplots(figsize=(25, 10))

plt.plot(df.referencia, df.gasolina_comum_preco_revenda_avg, label="Gasolinha Comum - Média", color='green')
plt.plot(df.referencia, df.gasolina_comum_preco_revenda_min, label="Gasolinha Comum - Preço Mínimo", color='blue')
plt.plot(df.referencia, df.gasolina_comum_preco_revenda_max, label="Gasolinha Comum - Preço Máximo", color='red')

plt.xlabel("Período de Referencia")
plt.ylabel("Variação de Preços (R$)")

plt.xticks(rotation=45)

# Agrupamento de Labels
loc = plticker.MultipleLocator(base=3) 
ax.xaxis.set_major_locator(loc)

plt.title("Variação de Preços da Gasolina Comum")
plt.legend()
plt.show();
```
