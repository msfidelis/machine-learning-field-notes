```python
df = pd.read_csv("https://economia-popular-delivery-content-indices.s3.amazonaws.com/inflacao/inflacao.csv")
```

```python
df.tail()
```

```python
df = df[df['ipca_variacao_alimentacao'] != 0]
```

```python
df.index = pd.to_datetime(df.referencia, format='%Y-%m')
```

```python
cols = ['referencia','ipca_variacao', 'ipca_variacao_alimentacao', 'ipca_variacao_habitacao', 'ipca_variacao_artigos_residencia', 'ipca_variacao_vestuario', 'ipca_variacao_transporte', 'ipca_variacao_saude', 'ipca_variacao_despesas_pessoais', 'ipca_variacao_educacao', 'ipca_variacao_comunicacao']
df_ipca = df[cols]
```

```python
df_ipca.head()
```

```python
df_ipca.tail()
```

```python
cols_acumulado = ['referencia','ipca_acumulado_ano', 'ipca_acumulado_ano_alimentacao', 'ipca_acumulado_ano_habitacao', 'ipca_acumulado_ano_artigos_residencia', 'ipca_acumulado_ano_vestuario', 'ipca_acumulado_ano_transporte', 'ipca_acumulado_ano_saude', 'ipca_acumulado_ano_despesas_pessoais', 'ipca_acumulado_ano_educacao', 'ipca_acumulado_ano_comunicacao']
df_ipca_acumulado = df[cols_acumulado]
```

```python
df_ipca_acumulado.head()
```

```python
df_ipca_acumulado.tail()
```

```python
plt.figure(figsize=(25, 25))
df_ipca.plot(x='referencia', kind='bar', stacked=True,
        title='Stacked - IPCA')
plt.show()
```

```python
plt.figure(figsize=(25, 8))
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao, label="IPCA - Indice Geral", color='blue')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_alimentacao, label="IPCA - Alimentação", color='orange')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_habitacao, label="IPCA - Habitação", color='green')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_artigos_residencia, label="IPCA - Art. Residencia", color='red')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_vestuario, label="IPCA - Vestuario", color='purple')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_transporte, label="IPCA - Transporte", color='brown')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_saude, label="IPCA - Saúde", color='pink')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_despesas_pessoais, label="IPCA - Despesas Pessoais", color='grey')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_educacao, label="IPCA - Educação", color='olive')
plt.plot(df_ipca.referencia, df_ipca.ipca_variacao_comunicacao, label="IPCA - Comunicação", color='cyan')
plt.xlabel("Período de Referencia")
plt.ylabel("Variação do IPCA")

plt.xticks(rotation=90)

plt.title("IPCA - Decomposição por setor")
plt.legend()
plt.show()
```
