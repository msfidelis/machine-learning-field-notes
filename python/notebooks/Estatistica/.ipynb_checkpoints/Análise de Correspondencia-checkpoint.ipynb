# Exemplo de Análise de Correspondência Simples Multipla

## Setup do Projeto

```python
import pandas as pd
import numpy as np
import prince
import scipy.stats as stats 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/superheroes_v3.csv")
```

```python
df.head()
```

## Tratamento dos Dados

```python
df = df[['Creator', 'Species', 'Alignment']].copy()
```

* Criado um Dataset somente com as categorias que serão analisadas

```python
df.head()
```

### Tabelas de Frequencia para Conhecimento dos Dados

```python
df.Creator.value_counts()
```

```python
df.Alignment.value_counts()
```

```python
df.Species.value_counts()
```

```python
df.isnull().sum()
```

### Tratando Convertendo as Strings `None` para `NaN`

```python
df = df.replace('None', np.nan, regex=True)
```

```python
df = df.replace('-', np.nan, regex=True)
```

```python
df.isnull().sum()
```

### Gerando a Tabela de Contingência

```python
tbl_contigencia = pd.crosstab(df.Creator, df.Alignment)
```

```python
tbl_contigencia
```

### Realizando o teste chi-quadrado

```python
chi2, pvalor, graus_liberdade, valores_esperados = stats.chi2_contingency(tbl_contigencia)
```

```python
print(f'chi2 = {chi2:.2f} e p-valor = {pvalor:}')
```

* Adotando o intervalo de confiança padrão de 95%, também interpretado como nível de significância de 5%, podemos considerar p-valores menores que 0.05 rejeitariam a hipotese nula do teste chi-quadrado. 
* como nosso p-valor foi muito maior, podemos prosseguir com a análise

## Análise de Correspondência Simples

* Realizaremos o teste baseando no autor e alinhamento de seus personagens
* O objetivo é identificar uma tendência de um autor ao criar personagens de determinados alinhamentos

```python
df_alinhamento = df[['Creator', 'Alignment']].copy()
```

```python
df_alinhamento.head()
```

```python
df_alinhamento = df_alinhamento.dropna()
```

```python
df_alinhamento.head()
```

* Dropamos todas as linhas que possuiam valores nulos

```python
df_alinhamento.Alignment.value_counts()
```

```python
sns.countplot(data=df_alinhamento, x="Alignment", order=df_alinhamento.Alignment.value_counts().iloc[:3].index);
```

```python
df_alinhamento.Creator.value_counts()
```

```python
sns.countplot(data=df_alinhamento, y="Creator", order=df_alinhamento.Creator.value_counts().iloc[:20].index);
```

```python
tbl_contigencia_alinhamento = pd.crosstab(df_alinhamento.Creator, df_alinhamento.Alignment)
```

```python
# Num. de dimensões
ndim = (np.array(tbl_contigencia_alinhamento.shape) - 1).min()
# Criando o modelo ANACOR
ca = prince.CA(
    n_components=ndim,
    engine='auto',
)
ca.fit(tbl_contigencia_alinhamento);
```

```python
ca.plot_coordinates(
    tbl_contigencia_alinhamento, 
    figsize=(12, 8)
)
plt.title("Correspondência entre Alinhamento e Criador")
plt.show()
```

* Podemos identificar que criadores como `Netflix`, `Mortal Kombat`, `Konami`, tem maior tendência a criarem personagens de lalinhamento `Bad`
* `Valve`, `Nintendo` tem maiores tendências a criarem personagens de linhamento `Neutral`
* `Tolkien`, `Studios Ghibli`, `Disney` e `Pixar` possuem tendências de criar personagens de alinhamento `Good`

* Top 20 Criadores

## Análise de Correspondência Multipla

```python
df_alinhamento_especie = df[['Creator', 'Alignment', 'Species']].copy()
```

```python
df_alinhamento_especie = df_alinhamento_especie.dropna()
```

```python
import prince
mca = prince.MCA(
     n_components=2,
     n_iter=3,
     copy=True,
     check_input=True,
     engine='auto',
     random_state=42
)
mca = mca.fit(df_alinhamento_especie)
```

```python
mca.plot_coordinates(
     X=df_alinhamento_especie,
     ax=None,
     figsize=(6, 6),
     show_row_points=True,
     row_points_size=10,
     show_row_labels=False,
     show_column_points=True,
     column_points_size=30,
     show_column_labels=False,
     legend_n_cols=1
);
```
