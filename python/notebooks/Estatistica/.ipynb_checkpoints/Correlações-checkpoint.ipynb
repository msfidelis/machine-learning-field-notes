```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy
```

```python
url = "https://raw.githubusercontent.com/msfidelis/indices-economicos/main/data/inflacao/inflacao.csv"
inflacao_raw = pd.read_csv(url)
inflacao_raw.head()
```

```python
inflacao_raw.info()

## Análise Prévia do IPCA - Indice de Preços no Consumidor Amplo

ipca_atual_df = inflacao_raw[inflacao_raw['ano'] >= 2020]
ipca_atual_df.head()


x = ipca_atual_df.referencia
y = ipca_atual_df.ipca_variacao

plt.figure(figsize=(25, 8))
plt.plot(x, y, label="Variação do IPCA", color='purple')
plt.xlabel("Período de Referencia")
plt.ylabel("Variação do IPCA")

plt.xticks(x, rotation=45)

plt.title("Variação do IPCA - 2020 até agora")
# plt.legend()
plt.show()

inflacao_2000 = inflacao_raw[inflacao_raw['ano'] >= 2000]
consolidado_df = inflacao_2000[inflacao_2000['consolidado_ano'] == True]
consolidado_df.head()

x = consolidado_df.ano
y = consolidado_df.ipca_acumulado_ano

plt.figure(figsize=(25, 8))
plt.bar(x, y, label="Acumulo anual do IPCA", color='purple')
plt.xlabel("Período de Referencia")
plt.ylabel("Acumulo Anual do IPCA %")

plt.xticks(x, rotation=0)


plt.title("Acumulo do IPCA - 2000 até agora")
plt.show()

x = consolidado_df.ano
y = consolidado_df.inpc_acumulado_doze_meses

plt.figure(figsize=(25, 8))
plt.bar(x, y, label="Acumulo do IPCA nos Ultimos 12 meses", color='green')
plt.xlabel("Período de Referencia")
plt.ylabel("Acumulo do IPCA em 12 meses %")

plt.xticks(x, rotation=0)


plt.title("Acumulo do IPCA - 2000 até agora")
plt.show()

## Análise Prévia do INPC - Indice de Nacional de Preços ao Consumidor

inpc_atual_df = inflacao_raw[inflacao_raw['ano'] >= 2020]
print(inpc_atual_df.head())


x = inpc_atual_df.referencia
y = inpc_atual_df.inpc_variacao

plt.figure(figsize=(25, 8))
plt.plot(x, y, label="Variação do INPC", color='blue')
plt.xlabel("Período de Referencia")
plt.ylabel("Variação do INPC")

plt.xticks(x, rotation=45)


plt.title("Variação do INPC - 2020 até agora")
# plt.legend()
plt.show()

inflacao_2000 = inflacao_raw[inflacao_raw['ano'] >= 2000]
consolidado_df = inflacao_2000[inflacao_2000['consolidado_ano'] == True]
print(consolidado_df.head())

x = consolidado_df.ano
y = consolidado_df.inpc_acumulado_ano

plt.figure(figsize=(25, 8))
plt.bar(x, y, label="Acumulo anual do INPC", color='blue')
plt.xlabel("Período de Referencia")
plt.ylabel("Acumulo Anual do INPC %")

plt.xticks(x, rotation=0)


plt.title("Acumulo do INPC - 2000 até agora")
plt.show()

x = consolidado_df.ano
y = consolidado_df.inpc_acumulado_doze_meses

plt.figure(figsize=(25, 8))
plt.bar(x, y, label="Acumulo do INPC nos Ultimos 12 meses", color='blue')
plt.xlabel("Período de Referencia")
plt.ylabel("Acumulo do INPC em 12 meses %")

plt.xticks(x, rotation=0)


plt.title("Acumulo do INPC - 2000 até agora")
plt.show()

## Correlações entre a Variação de IPCA e INPC

df_corr_variacao = inflacao_raw[["referencia", "ipca_variacao", "inpc_variacao"]]
df_corr_variacao.head()

 corr = df_corr_variacao.corr()
 corr.style.background_gradient(cmap='coolwarm')

### Correlação com Pandas

print("Pandas - Pearson")
df_corr_variacao.ipca_variacao.corr(df_corr_variacao.inpc_variacao, method="pearson")    

print("Pandas - Spearman")
df_corr_variacao.ipca_variacao.corr(df_corr_variacao.inpc_variacao, method="spearman")    

print("Pandas - Kendall")
df_corr_variacao.ipca_variacao.corr(df_corr_variacao.inpc_variacao, method="kendall")    

### Correlação com SciPy

print("SciPy - Pearson")

scipy.stats.pearsonr(
    df_corr_variacao.ipca_variacao, 
    df_corr_variacao.inpc_variacao
    )

print("SciPy - Spearman")

scipy.stats.spearmanr(
    df_corr_variacao.ipca_variacao, 
    df_corr_variacao.inpc_variacao
    )

print("SciPy - Kendall")

scipy.stats.kendalltau(
    df_corr_variacao.ipca_variacao, 
    df_corr_variacao.inpc_variacao
    )

print("SciPy - Linear Regression")

result = scipy.stats.linregress(
    df_corr_variacao.ipca_variacao, 
    df_corr_variacao.inpc_variacao
    )

print(result)

print("SciPy - Linear Regression / PLOT")

slope, intercept, r, p, stderr = scipy.stats.linregress(
    df_corr_variacao.ipca_variacao, 
    df_corr_variacao.inpc_variacao
    )

line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'

print(result)

### Visualização de Correlação

plt.rcParams.update({'figure.figsize':(25,8), 'figure.dpi':100})
sns.lmplot(x="ipca_variacao", y="inpc_variacao", data=df_corr_variacao);

x = df_corr_variacao.ipca_variacao
y = df_corr_variacao.inpc_variacao

plt.figure(figsize=(25, 8))
plt.scatter(x, y, c=y, cmap='Spectral')
plt.colorbar()
plt.title("Correlação entre IPCA e INPC")
plt.show()
```
