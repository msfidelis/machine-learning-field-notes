```python
fig, ax = plt.subplots(figsize=(25, 10))

plt.plot(df.referencia, df.gasolina_comum_preco_revenda_avg, label="Gasolinha Comum - MÃ©dia", color='green')
plt.plot(df.referencia, df.gasolina_comum_preco_revenda_min, label="Gasolinha Comum - PreÃ§o MÃ­nimo", color='blue')
plt.plot(df.referencia, df.gasolina_comum_preco_revenda_max, label="Gasolinha Comum - PreÃ§o MÃ¡ximo", color='red')

plt.xlabel("PerÃ­odo de Referencia")
plt.ylabel("VariaÃ§Ã£o de PreÃ§os (R$)")

plt.xticks(rotation=45)

# Agrupamento de Labels
loc = plticker.MultipleLocator(base=3) 
ax.xaxis.set_major_locator(loc)

plt.title("VariaÃ§Ã£o de PreÃ§os da Gasolina Comum")
plt.legend()
plt.show();
```
