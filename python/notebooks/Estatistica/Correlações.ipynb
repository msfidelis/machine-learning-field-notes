# Correlação de Pearson

O coeficiente de correlação de pearson testa a correlação entre duas variáveis métricas. 
Sejam duas variáveis X e Y, ambas quantitativas, preferencialmente contínuas. 

A correlação de Pearson avalia a relação linear entre duas variáveis contínuas. 
Uma relação é linear quando a mudança em uma variável é associada a uma udança proporcional na outra variável.

Por exemplo, você poderia usar uma correlação de Pearson para avaliar se **aumentos na temperatura da instalação de produção estão associados a uma redução da espessura da cobertura de chocolate**

#### Utilizado quando: 
* Utilizado quando: Temos normalidade dos dados, e eles são LINEARES

### Testando a Correlação entre IPCA15 e IPCA

### Pandas - Pearson

#### Eliminando as linhas dos periodos nos quais o IPCA15 não era contabilizado

O ipca15 começou a ser contabilizado como série histórica a partir do ano 2000, enquanto sua série principal existe desde os anos 80. Para isso precisamos remover todos os periodos nos quais não temos a estimativa do ipca15 para realizar a comparacão. 

```python
df_ipca = inflacao_raw[inflacao_raw['ipca15_variacao'] != 0]
```

```python
df_ipca.head()
```

```python
df_ipca.ipca_variacao.corr(df_ipca.ipca15_variacao, method="pearson")
```

### Utilizando o Scipy para testes de correlacão. 

A Biblioteca scipy devolve além do coefiente de correlacão, o p-valor da estatística

##### P-VALOR

Em Suma: P-VALOR
Aceita-se um valor de corte de P-valor em 5%, ou 0.05
* Se o p-valor for menor que **0.05**, ou **5**, pode-se dizer que a estatistica "deu certo"
* Se o p-valor for maior que **0.05**, ou **5**, pode-se dizer que a estatística "deu errado"


```python
scipy.stats.pearsonr(
    df_ipca.ipca_variacao, 
    df_ipca.ipca15_variacao
    )
```

# Correlação de Spearman

A correlação de Spearman avalia a relação monotônica entre duas variáveis contínuas ou ordinais. 
Em uma relação monotônica, as variáveis tendem a mudar juntas mas não necessariamente a uma taxa constante. 

O coeficiente de correlação de Spearman baseia-se nos valores classificados de cada variável, em vez de os dados brutos.

A correlação de Spearman é muito usada para avaliar relações envolvendo variáveis ordinais. Por exemplo, você poderia usar a correlação de Spearman para avaliar se a ordem na qual os funcionários executam um teste está relacionada ao número de meses de emprego.

#### Utilizado quando
* Amostras Maiores

## Testando a Correlação entre o INCC e o IPA 

```python
df_incc = inflacao_raw[inflacao_raw['incc_variacao'] != 0]
```

```python
df_incc.describe()
```

```python
scipy.stats.spearmanr(
    df_incc.ipa_variacao, 
    df_incc.incc_variacao
    )
```

# Correlação de Kendall

O coeficiente de correlação Tau de Kendall serve para verificar se existe correlação entre duas variáveis ordinais. É um método adequado quando amostras têm tamanhos reduzidos, pois o método é mais preciso. 

E pode ser estendido a correlações parciais, quando o efeito de uma terceira variável, que age sobre X e Y, é retirado antes de determinar se X e Y estão relacionadas.

Coeficiente de Kendall é, muitas vezes, interpretado como uma medida de concordância entre dois conjuntos de classificações relativas a um conjunto de objetos de estudo


#### Utilizado quando

* Parecido com Spearman, mas utilizado para amostras menores

* Os números ordinais representam posições em uma determinada sequência, como: primeiro, segundo, décimo quinto, entre outros.

```python
scipy.stats.kendalltau(
    df_incc.ipa_variacao, 
    df_incc.incc_variacao
    )
```
